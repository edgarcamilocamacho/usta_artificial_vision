# Cargando imágenes con Python

Algunas librerías:

* *numpy*: Computación matricial
* *matplotlib*: Gráficas
* *cv2*: OpenCV, Procesamiento de imágenes
* *imageio*: Leer y escribir imágenes

Comando de instalación:

```
$ pip install opencv-python numpy matplotlib imageio
```

````python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread
import cv2
%matplotlib inline
````

## Leer imágenes con *imageio*

Leer una imagen y mostrarla con *matplotlib*:

````python
img = imread('./data/img/cat1.jpg')
plt.imshow(img)
plt.show()
````

Las imágenes se almacenan como arreglos de *numpy* (imageio.core.util.Array es casi lo mismo):

````python
type(img)
````

Es una matrix tridimencional, 400 filas, 248 columnas y 3 canales.

````python
img.shape
````

Por defecto, las imágenes se leen en entero de 8 bits:

````python
img.dtype
````

Los arreglos pueden ser indexados de difetentes maneras:

RGB de un píxel:

````python
print(img[10,20])
````

Valor G de un píxel:

````python
print(img[15,141,1])
````

Una subsección del canal B:

````python
print(img[15:25,108:112,2])
````

Asimismo, se pueden graficar porciones de la imagen:

````python
plt.imshow(img[50:150,50:150,:])
plt.show()
````

## Canales independientes

Rojo

````python
fig=plt.figure(figsize=(10,10), dpi= 70, facecolor='w', edgecolor='k')

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Imagen Original RGB')

img_red_rgb = np.uint8(img * [1.0, 0.0, 0.0])
plt.subplot(1, 3, 2)
plt.imshow(img_red_rgb)
plt.title('Imagen RGB, con canales\nverde y azul "apagados"')

img_red_gray = img[:,:,0]
plt.subplot(1, 3, 3)
plt.imshow(img_red_gray, cmap='gray')
plt.title('Canal rojo presentado\ncomo intensidades')

plt.show()
````

Verde

````python
fig=plt.figure(figsize=(10,10), dpi= 70, facecolor='w', edgecolor='k')

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Imagen Original RGB')

img_red_rgb = np.uint8(img * [0.0, 1.0, 0.0])
plt.subplot(1, 3, 2)
plt.imshow(img_red_rgb)
plt.title('Imagen RGB, con canales\nrojo y azul "apagados"')

img_red_gray = img[:,:,1]
plt.subplot(1, 3, 3)
plt.imshow(img_red_gray, cmap='gray')
plt.title('Canal verde presentado\ncomo intensidades')

plt.show()
````

Azul

````python
fig=plt.figure(figsize=(10,10), dpi= 70, facecolor='w', edgecolor='k')

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Imagen Original RGB')

img_red_rgb = np.uint8(img * [0.0, 0.0, 1.0])
plt.subplot(1, 3, 2)
plt.imshow(img_red_rgb)
plt.title('Imagen RGB, con canales\nrojo y verde "apagados"')

img_red_gray = img[:,:,2]
plt.subplot(1, 3, 3)
plt.imshow(img_red_gray, cmap='gray')
plt.title('Canal azul presentado\ncomo intensidades')

plt.show()
````

## "Enrojeciendo" imagen

````python
fig=plt.figure(figsize=(7,7), dpi= 70, facecolor='w', edgecolor='k')

img_tinted = img * [1, 0.9, 0.8]

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Imagen Original')
                    
plt.subplot(1, 2, 2)
plt.imshow(np.uint8(img_tinted))
plt.title('Imagen Enrojecida')
                    
plt.show()
````

# Leyendo imagen con OpenCV

````python
img = cv2.imread('./data/img/cat1.jpg')
plt.imshow(img)
plt.show()
````

Las imagenes en OpenCV son leídas como BGR, mas no RGB, como *matplotlib* lo espera. Se pueden invertir los canales así:

````python
img_rgb = img[:,:,[2,1,0]]
# img_rgb = img[...,::-1]  # o así
plt.imshow(img_rgb)
plt.show()
````

## Convertir de RGB a escala de grises

````python
img = cv2.imread('./data/img/cat1.jpg')
print(f'img.shape = {img.shape}')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(f'img_gray.shape = {img_gray.shape}')
plt.imshow(img_gray, cmap = 'gray')
plt.show()
````

## Grayscale Images, importing it directly.

````python
img = cv2.imread('./data/img/cat1.jpg', 0)
print(f'img.shape = {img.shape}')
plt.imshow(img_gray, cmap = 'gray')
plt.show()
````
